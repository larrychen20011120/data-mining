```python
import matplotlib.pyplot as plt
import os
```

```python
entry = "Clustering_testdata"

index = 0

data_results = dict()

for filename in os.listdir(entry):

    if filename.endswith(".out"):

        clusters = dict()

        with open(os.path.join(entry, filename), "r") as f:
            for line in f.readlines():
                line = line.rstrip().split()
                x, y, cluster = float(line[0]), float(line[1]), int(line[2])
                if cluster not in clusters:
                    clusters[cluster] = [[], []]

                clusters[cluster][0].append(x)
                clusters[cluster][1].append(y)

        data_results[filename.split(".")[0]] = clusters

len(data_results)
```

```python
plt.figure(figsize=(10, 6))
cnt = 1

for name, clusters in data_results.items():
    plt.subplot(int(f"23{cnt}"))
    plt.title(name)
    
    for cluster, data in clusters.items():    
        plt.scatter(data[0], data[1], label=cluster)
        
    cnt += 1

plt.tight_layout()
plt.show()
```
